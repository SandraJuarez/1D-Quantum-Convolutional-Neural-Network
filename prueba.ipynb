```python
import jax 
import jax.numpy as jnp
```

```python
# Set the random seed
key = jax.random.PRNGKey(0)

# Generate a vector of 20 random numbers from a uniform distribution
random_vector = jax.random.uniform(key, shape=(21,))

print(random_vector)
```

```python
import jax.numpy as jnp
import jax.lax as lax

# Creamos un arreglo de ejemplo
x = jnp.arange(30).reshape(3, 10)
print(x)

# Extraer desde la posición 0 a la 5 en la segunda dimensión
sliced = lax.slice(x, (0, 2), (3, 7))  # Cortamos de (0,2) a (3,7)
print(sliced)
```
