```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
loss_4q=torch.load('mackey/Lossmackeystrongly_layers24no_reupload.pt')
loss_6q=torch.load('mackey/Lossmackeystrongly_layers24.pt')
loss_8q=torch.load('mackey/Lossmackeystrongly_layers42.pt')
epochs=30
```

```python
print(loss_8q)
```

```python
plt.plot(np.arange(1,epochs+1),loss_4q , 'red', label='Loss_nr')
plt.plot(np.arange(1,epochs+1),loss_6q , 'blue', label='Loss_r')
plt.plot(np.arange(1,epochs+1),loss_8q , 'green', label='Loss_42')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
plt.title('Loss')
plt.show()
```

```python
loss_sl=torch.load('legendre3/Losslegendre3strongly_layers42.pt')
loss_bl=torch.load('legendre3/Losslegendre3basic_layers42.pt')
loss_rl=torch.load('legendre3/Losslegendre3random_layers42.pt')
loss_cl=torch.load('legendre3/Losslegendre3custom_layers42.pt')
epochs=30
```
